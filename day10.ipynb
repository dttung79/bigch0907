```python
import matplotlib.pyplot as plt
import numpy as np

# create 100 linearly spaced numbers from -10 to 10
x = np.linspace(-10, 10, 10)
y = 3*x**2 -2*x + 1 # y = 3x^2 - 2x + 1

# draw the plot
plt.plot(x, y, 'vb')

# set the x axis label
plt.xlabel('x')
# set the y axis label
plt.ylabel('y')

# set the plot title
plt.title('y = 3x^2 - 2x + 1')

# show the plot
plt.show()
```

```python
x = np.linspace(-1, 3, 100)
y = 3*x**2

plt.xlabel('x')
plt.ylabel('y')

plt.title('y = 3x^2')

plt.plot(x, y, 'r')
plt.show()
```
